```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
import yfinance as yf

lookback = 60 
feature_columns = ['Open', 'High', 'Low', 'Close', 'Volume']


def fetch_data(ticker_list):
    all_data = []
    for ticker in ticker_list:
        try:
            stock_data = yf.download(ticker, period="max") 
            first_date = stock_data.index.min()
    
            ideal_start_date = pd.to_datetime('2010-01-01') 
            start_date = ideal_start_date if first_date < ideal_start_date else first_date
            data = yf.download(ticker, start= start_date, end="2025-01-01")
            data = data[feature_columns].dropna()
            data['ticker'] = ticker
            all_data.append(data)
            
        except Exception as e:
            print(f"Error fetching {ticker}: {e}")
    return pd.concat(all_data)


def preprocess_combined_data(data):
    scalers = {}
    enc = OneHotEncoder(sparse=False)
    tickers_encoded = enc.fit_transform(data[['ticker']])

    all_sequences = []
    all_targets = []
    for ticker in data['ticker'].unique():
        stock_data = data[data['ticker'] == ticker]
        stock_features = stock_data[feature_columns]
        scaler = MinMaxScaler()
        scaled_data = scaler.fit_transform(stock_features)
        scalers[ticker] = scaler  

        for i in range(lookback, len(scaled_data)):
            seq = scaled_data[i - lookback:i]
            seq_with_ticker = np.hstack([seq, tickers_encoded[i - lookback]]) 
            all_sequences.append(seq_with_ticker)
            all_targets.append(scaled_data[i, 3])  

    return np.array(all_sequences), np.array(all_targets), scalers


def create_model(input_shape, num_tickers):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50))
    model.add(Dropout(0.2))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mse')
    return model


ticker_list = list(set(final_ticker_list))
print("Fetching data...")
data = fetch_data(ticker_list)

print("Preprocessing data...")
X, y, scalers = preprocess_combined_data(data)

print("Training model...")
model = create_model((lookback, X.shape[2]), len(ticker_list))
model.fit(X, y, epochs=10, batch_size=32, verbose=1)

model.save("stock_lstm_model.h5")
```
