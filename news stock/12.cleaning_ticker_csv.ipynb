```python
from pyspark.sql import SparkSession
import pyspark
```

```python
spark = SparkSession.builder\
                    .appName("ticker_csv_cleaning_process")\
                        .getOrCreate()
```

```python
spark
```

```python
df = spark.read.format("csv").option("header", True).option("inferSchema", "true").load("all_yahoo_tikckers.csv")
```

```python
from pyspark.sql import functions as F
```

```python
india_stock_ticker_df = df.filter(F.col("Country") == "India")
```

```python
india_stock_ticker_df = india_stock_ticker_df.drop(F.col("Country"))
```

```python
india_stock_ticker_df.show()
```

```python
india_stock_ticker_df.collect()
```

```python
india_stock_ticker_df.write.format("csv").option("path","indian_stock_ticker.csv")
```

```python
# india_stock_ticker_df.write.csv("indian_stock_ticker.csv")
```

```python
only_stock_name_df = india_stock_ticker_df.select(F.col("Name"))
```

```python
# Convert the DataFrame to a Pandas DataFrame
pandas_df = only_stock_name_df.toPandas()

# Convert the "name" column to a list
name_list = pandas_df["Name"].tolist()

print(name_list)
```
